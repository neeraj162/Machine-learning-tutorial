```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Y=pd.read_csv('test.csv')
Y.shape
```

```python
z1 = pd.DataFrame(Y.isnull().sum())
z1[z1[0]>0]
```

```python
Y.drop(Y[['Alley','PoolQC','Fence','MiscFeature']],axis=1,inplace=True)
np.array(Y.isnull().sum())
```

```python
Y.LotFrontage.fillna(Y.LotFrontage.mean(),inplace=True)

Y.MasVnrType.fillna(Y.MasVnrType.mode()[0],inplace=True)
Y.MasVnrArea.fillna(Y.MasVnrArea.mean(),inplace=True)

Y.BsmtCond.fillna(Y.BsmtCond.mode()[0],inplace=True)
Y.BsmtQual.fillna(Y.BsmtQual.mode()[0],inplace=True)

Y.BsmtExposure.fillna(Y.BsmtExposure.mode()[0],inplace=True)

Y.BsmtFinType1.fillna(Y.BsmtFinType1.mode()[0],inplace=True)
Y.BsmtFinType2.fillna(Y.BsmtFinType2.mode()[0],inplace=True)

Y.FireplaceQu.fillna(Y.FireplaceQu.mode()[0],inplace=True)
Y.GarageType.fillna(Y.GarageType.mode()[0],inplace=True)

Y.GarageYrBlt.fillna(Y.GarageYrBlt.mean(),inplace=True)

Y.GarageFinish.fillna(Y.GarageFinish.mode()[0],inplace=True)
Y.GarageQual.fillna(Y.GarageQual.mode()[0],inplace=True)
Y.GarageCond.fillna(Y.GarageCond.mode()[0],inplace=True)
```

```python
Y.MSZoning.fillna(Y.MSZoning.mode()[0],inplace=True)
Y.Utilities.fillna(Y.Utilities.mode()[0],inplace=True)
Y.Exterior2nd.fillna(Y.Exterior2nd.mode()[0],inplace=True)
Y.Exterior1st.fillna(Y.Exterior1st.mode()[0],inplace=True)

Y.BsmtFinSF1.fillna(Y.BsmtFinSF1.mean(),inplace=True)
Y.BsmtFinSF2.fillna(Y.BsmtFinSF2.mean(),inplace=True)
Y.BsmtUnfSF.fillna(Y.BsmtUnfSF.mean(),inplace=True)
Y.TotalBsmtSF.fillna(Y.TotalBsmtSF.mean(),inplace=True)
Y.BsmtFullBath.fillna(Y.BsmtFullBath.mean(),inplace=True)
Y.BsmtHalfBath.fillna(Y.BsmtHalfBath.mean(),inplace=True)
Y.GarageCars.fillna(Y.GarageCars.mean(),inplace=True)
Y.GarageArea.fillna(Y.GarageArea.mean(),inplace=True)

Y.KitchenQual.fillna(Y.KitchenQual.mode()[0],inplace=True)
Y.Functional.fillna(Y.Functional.mode()[0],inplace=True)
Y.SaleType.fillna(Y.SaleType.mode()[0],inplace=True)
```

```python
np.array(Y.isnull().sum())
Y.shape
```

```python
g = Y.columns.to_series().groupby(Y.dtypes).groups
a = {k.name: v for k, v in g.items()}
cat_col = list(a['object'])
lst=[]
for i in range(len(cat_col)):
    lst.append(len(Y[cat_col[i]].value_counts()))
lst
```
