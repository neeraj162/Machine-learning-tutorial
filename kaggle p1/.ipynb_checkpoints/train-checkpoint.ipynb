```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
X = pd.read_csv('train.csv')
X
```

```python
X.info()
```

```python
X.shape
z1 = pd.DataFrame(X.isnull().sum())
z1[z1[0]>0]
```

```python
X.drop(X[['Alley','PoolQC','Fence','MiscFeature']],axis=1,inplace=True)
np.array(X.isnull().sum())
```

```python
X.LotFrontage.fillna(X.LotFrontage.mean(),inplace=True)

X.MasVnrType.fillna(X.MasVnrType.mode()[0],inplace=True)
X.MasVnrArea.fillna(X.MasVnrArea.mean(),inplace=True)

X.BsmtCond.fillna(X.BsmtCond.mode()[0],inplace=True)
X.BsmtQual.fillna(X.BsmtQual.mode()[0],inplace=True)

X.BsmtExposure.fillna(X.BsmtExposure.mode()[0],inplace=True)

X.BsmtFinType1.fillna(X.BsmtFinType1.mode()[0],inplace=True)
X.BsmtFinType2.fillna(X.BsmtFinType2.mode()[0],inplace=True)

X.FireplaceQu.fillna(X.FireplaceQu.mode()[0],inplace=True)
X.GarageType.fillna(X.GarageType.mode()[0],inplace=True)
X.Electrical.fillna(X.Electrical.mode()[0],inplace=True)

X.GarageYrBlt.fillna(X.GarageYrBlt.mean(),inplace=True)

X.GarageFinish.fillna(X.GarageFinish.mode()[0],inplace=True)
X.GarageQual.fillna(X.GarageQual.mode()[0],inplace=True)
X.GarageCond.fillna(X.GarageCond.mode()[0],inplace=True)

```

```python
np.array(X.isnull().sum())
```

```python
X
```

```python
g = X.columns.to_series().groupby(X.dtypes).groups
a = {k.name: v for k, v in g.items()}
cat_col = list(a['object'])
lst=[]
for i in range(len(cat_col)):
    lst.append(len(X[cat_col[i]].value_counts()))
lst
```

```python
Y1 = pd.read_csv('test1.csv')
newdf = pd.concat([X,Y1],axis=0)
newdf
```

```python
def handling(lst):
    df1=newdf
    for i in range(len(lst)):
        dummies = pd.get_dummies(newdf[lst[i]],prefix=lst[i])
        newdf.drop([lst[i]], axis=1,inplace=True)
        if i==0:
            df1 = dummies.copy()
        else:
            df1 = pd.concat([df1,dummies],axis=1)
    df1 = pd.concat([newdf,df1],axis=1)
    return df1
```

```python
df2 = handling(cat_col)
np.array(df2.columns)
df2
```

```python
df2 = df2.loc[:,~df2.columns.duplicated()]
df2
```

```python
X = df2.iloc[:1460,:]
Y = df2.iloc[1460:,:]
X_train_main = X.drop('SalePrice',axis=1)
Y_train_main = X['SalePrice']
X_test_main = Y.drop('SalePrice',axis=1)
X_test_main
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

model_params = {
    'linear': {
        'model': LinearRegression(),
        'params' : {
            
        }  
    },
    'random_forest': {
        'model': RandomForestRegressor(),
        'params' : {
            'criterion': ['mse', 'mae']
        }
    },
    'decision_tree' : {
        'model': DecisionTreeRegressor(),
        'params': {
            'criterion': ['mse', 'friedman_mse', 'mae', 'poisson']
        }
    }
}


from sklearn.model_selection import GridSearchCV
import pandas as pd
scores = []

for model_name, mp in model_params.items():
    clf =  GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    clf.fit(X_train_main, Y_train_main)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
    
df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(criterion='mae')
model.fit(X_train_main,Y_train_main)
```

```python
y_pred = model.predict(X_test_main)
y_pred
```

```python
final_df = pd.DataFrame(X_test_main['Id'])
final_df['SalePrice'] = y_pred
final_df
```

```python
final_df.to_csv('predictions.csv',index=False)
```
