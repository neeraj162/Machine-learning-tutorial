```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
newdf=pd.read_csv('country_wise_latest.csv')
newdf.replace([np.inf, -np.inf], np.nan, inplace=True).dropna(subset=["col1", "col2"], how="all")
newdf.iloc[:,10]
```

```python
X=np.array(newdf.iloc[:,0].head(10))
Y=np.array(newdf.iloc[:,2].head(10))
```

```python
%matplotlib inline
plt.plot(X,Y,'g')
plt.title("country and death rates")
plt.xlabel("Countries")
plt.ylabel("Deaths")
plt.show()
```

```python
newdf.to_csv('country_wise_latest.csv',index=False)
```
